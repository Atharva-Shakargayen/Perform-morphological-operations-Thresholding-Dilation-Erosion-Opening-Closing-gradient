```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path='/content/imaageeeeee.png'
img=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)


## Threshold

_,binary_image=cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)

## Define kernel
kernel=np.ones((5,5),np.uint8)



#1. Dilation
dilation=cv2.dilate(binary_image,kernel,iterations=1)

#2. Erosion
erosion=cv2.erode(binary_image,kernel,iterations=1)

#3. Closing
closing=cv2.morphologyEx(binary_image,cv2.MORPH_CLOSE,kernel)

#4. Opening
opening=cv2.morphologyEx(binary_image,cv2.MORPH_OPEN,kernel)

#5. Gradient
gradient=cv2.morphologyEx(binary_image,cv2.MORPH_GRADIENT,kernel)

#6. Top hat
tophat=cv2.morphologyEx(binary_image,cv2.MORPH_TOPHAT,kernel)

#7. Blackhat
blackhat=cv2.morphologyEx(binary_image,cv2.MORPH_BLACKHAT,kernel)


##Display
titles=['Original','Dilation','Erosion','Closing','Opening','Gradient','TopHat','BlackHat']
images=[img,dilation,erosion,closing,opening,gradient,tophat,blackhat]

#Plotting
plt.figure(figsize=(15,15))
for i in range(8):
  plt.subplot(3,3,i+1)
  plt.imshow(images[i],cmap='gray')
  plt.title(titles[i])
  plt.axis('off')
plt.tight_layout()
plt.show()
```
